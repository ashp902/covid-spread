```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../imputed-data.csv', index_col = 0)
data
```

```python
data.info()
```

```python
data = data.drop(columns=['county_fips', 'state_fips', 'state_name', 'latitude', 'longitude'])
```

```python
cat_columns = data.select_dtypes(['object']).columns
data[cat_columns] = data[cat_columns].apply(lambda x: pd.factorize(x)[0])
data
```

```python
data.county_name.unique()
```

```python
grouped = data.groupby(by=['county_name'])
```

```python
dataframes = []
for county_name in data.county_name.unique() :
    dataframes.append(grouped.get_group(county_name))
```

```python
len(dataframes)
```

```python
import pickle

with open('dataframes.pkl', 'wb') as file :
    pickle.dump(dataframes, file)
```

```python
with open('dataframes.pkl', 'rb') as file :
    x = pickle.load(file)
print(len(x))
```
